1. Money Change Again


```python
%%writefile MoneyChange.cpp
#include <iostream>
#include <vector>
using namespace std;

long number_coins(int money) {
    vector<long> min_coins(money + 1);
    int Num_coins = 0;
    min_coins[0] = 0;
    for(int m=1;m<=money;m++){
        min_coins[m] = 2147483647;
        if(m>=1){
            Num_coins = min_coins[m-1]+1;
            if(Num_coins<min_coins[m]){
                min_coins[m] = Num_coins;
            }
        }
        if(m>=3){
            Num_coins = min_coins[m-3]+1;
            if(Num_coins<min_coins[m]){
                min_coins[m] = Num_coins;
            }
        }
        if(m>=4){
            Num_coins = min_coins[m-4]+1;
            if(Num_coins<min_coins[m]){
                min_coins[m] = Num_coins;
            }
        }
    }
    return min_coins[money];
}

int main() {
    int n;
    cin >> n;
    cout << number_coins(n) << "\n";
    return 0;
}
```

2. Primitive Calculator

```python
%%writefile PrimitiveCalculator.cpp

#include <iostream>
#include <vector>
#include <iterator>
#include <algorithm>
#include <array>

using namespace std;

    vector<int> optimal_sequence(int n) {
        
      vector<int> sequence;

      int arr[n + 1];

      for (int i = 1; i < (n+1); i++) {
          arr[i] = arr[i - 1] + 1;
          if (i % 2 == 0){ arr[i] = min(1 + arr[i / 2], arr[i]);}
          if (i % 3 == 0){ arr[i] = min(1 + arr[i / 3], arr[i]);}

      }

      for (int i = n; i > 1;) {
          sequence.push_back(i);
          if (arr[i - 1] == arr[i] - 1){
              i = i - 1;
          }else if (i % 2 == 0 && (arr[i / 2] == arr[i] - 1)){
              i = i / 2;
          }else if (i % 3 == 0 && (arr[i / 3] == arr[i] - 1)){
              i = i / 3;
          }
      }
      sequence.push_back(1);

      reverse(sequence.begin(), sequence.end());
      return sequence;
   }

int main() {
        int n;
        cin >> n;
        vector<int> sequence;
        sequence = optimal_sequence(n);
        cout << sequence.size()-1 << endl;
        for (int i = 0; i < sequence.size(); i++)
        cout << sequence[i] << " ";
        
        return 0;
    }
```
