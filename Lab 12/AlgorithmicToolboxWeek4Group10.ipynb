### 1. Binary Search

```python
%%writefile binarySearch.cpp
#include <iostream>
using namespace std;

int binary_search(int a[],int low, int high, int b) {
    if(high<low){
        return -1;
    }
    int mid = low + (high-low)/2;
    if(b==a[mid]){
        return mid;
    }else if(b<a[mid]){
        return binary_search(a,low,mid-1,b);
    }else{
        return binary_search(a,mid+1,high,b);
    }
}

int main() {
    int n, k, b;
    cin >> n;
    int  a[n];
    for(int i=0;i<n;i++){
        cin >> a[i];
    }
    cin >> k;
    for(int i=0;i<k;i++){
        cin >> b;
        cout << binary_search(a,0,n-1,b) << " ";
    }
    return 0;
}
```

### 2. Majority Element

```python
%%writefile MajorityElement.cpp
#include <stdio.h>
#include <stdlib.h>
#include <iostream>
using namespace std;

int compare (const void * a, const void * b){
  return ( *(long long*)a - *(long long*)b );
}

int main (){
    long long n;
    cin >> n;
    long long  a[n];
    for(int i=0;i<n;i++){
        cin >> a[i];
    }
    qsort (a, n, sizeof(long long), compare);
    int m = n/2;
    bool bandera = false;
    for(int i=0;i<n-m;i++){
        if(a[i]==a[m+i]){
            cout << 1 << endl;
            bandera = true;
            break;
        }
    }
    if(!bandera) cout << 0 << endl;
    return 0;
}
```

### 3. Improving Quick Sort

```python
%%writefile ImprovingQuickSort.cpp
#include <iostream>
#include <stdio.h>
#include <stdlib.h>
#include <iostream>
using namespace std;

int compare (const void * a, const void * b){
  return ( *(long long*)a - *(long long*)b );
}

int main (){
    long long n;
    cin >> n;
    long long  a[n];
    for(int i=0;i<n;i++){
        cin >> a[i];
    }
    qsort (a, n, sizeof(long long), compare);
    for(int i=0;i<n;i++){
        cout << a[i] << " ";
    }
    return 0;
}
```
