### 3. Car Fueling


```python
%%writefile CarFueling.cpp
#include <iostream>
using namespace std;

int number_refills(int n, int m, int d, int stops[]) {
    int refill = 0, current_refill = 0, last_refill;
    while(current_refill<=n){
        last_refill = current_refill;
        while(current_refill<=n && stops[current_refill+1]-stops[last_refill]<=m){
            current_refill++;
        }
        if(current_refill==last_refill) return -1;
        if(current_refill<=n) refill++;
    }
    return refill;
}

int main() {
    int d,m,n;
    cin >> d;
    cin >> m;
    cin >> n;
    int stops[n+2];
    stops[0] = 0;
    for(int i=1;i<=n;i++){
        cin >> stops[i];
    }
    stops[n+1] = d;
    cout << number_refills(n,m,d,stops) << "\n";
    return 0;
}
```

```python
!g++ CarFueling.cpp -o CarFueling
!./CarFueling
```

### 6 Maximum Number of Prizes


```python
%%writefile MaximumNumberPrizes.cpp
#include <iostream>
using namespace std;

int portions(int k) {
    int i = 1, portions = 0;
    while(k>0){
        if(k-i>i){
            portions++;
            k = k - i;
            i++;
        }else{
            k = 0;
            portions++;
        }
    }
    return portions;
}

int main() {
    int k, porc = 0;
    cin >> k;
    porc = portions(k);
    cout << porc << "\n";
    int i = 1, j = 0;
    while(j<porc){
        if(k-i>i){
            cout << i << " ";
            k = k - i;
            i++;
        }else{
            cout << k << " ";
        }
        j++;
    }
    return 0;
}
```
